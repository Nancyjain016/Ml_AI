```python
vid = cv2.VideoCapture(0)
while True:
    flag, img = vid.read()
    if flag:
       cv2.imshow('Preview', img)
       key = cv2.waitKey(1)
       if key==ord('q'):
           break
    else:
        break
vid.release()
cv2.destroyAllWindows()       
```
